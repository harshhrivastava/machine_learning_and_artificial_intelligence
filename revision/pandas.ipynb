```python
import numpy as np
# Import Requirements

import seaborn as sns
```

```python
# Load Dataset

df = sns.load_dataset('penguins')
df
```

```python
# Get Information About The Dataset

df.info()
```

```python
# Get Description About The Numerical Columns Of The Dataset

df.describe()
```

```python
# Get Description About The Categorical Columns Of The Dataset

df.describe(include='object')
```

```python
# Check Missing Values In Each Column And Calculate Percentage

missing_values_count = df.isnull().sum()
percentage_missing_values_count = (missing_values_count * 100) / df.count(axis=0)
percentage_missing_values_count
```

```python
# Drop Rows With Missing Values

df = df.dropna()
df
```

```python
# Rename Column Names

df = df.rename(columns={
    'bill_length_mm': 'bill_length',
    'bill_depth_mm': 'bill_depth',
    'flipper_length_mm': 'flipper_length',
    'body_mass_g': 'body_mass',
})
df
```

```python
# Visualize the relation between flipper_length and body_mass

sns.lmplot(
    data=df,
    x='flipper_length',
    y='body_mass',
)
```

```python
# Visualize the relation between body_mass and species

sns.boxplot(
    data=df,
    x='species',
    y='body_mass',
)
```

```python
# Visualize the relation between body_mass and gender

sns.boxplot(
    data=df,
    x='sex',
    y='body_mass',
)
```

```python
# Visualize the relation between body_mass and island

sns.boxplot(
    data=df,
    x='island',
    y='body_mass',
)
```

```python
# Use Apply Function To The Sex Column

def format_gender(sex: str):
    if isinstance(sex, str):
        if sex == 'Male':
            return 0
        else:
            return 1
    return np.nan

df['sex'].apply(lambda i: format_gender(i))
```
